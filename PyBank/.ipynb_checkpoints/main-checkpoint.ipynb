```python
#Python-Challenge-1

#In this assignment, you will create a Python script that analyzes the financial records of your company.  
#Inside your starter code, you will find a financial dataset in the budget_data.csv file. 
#This dataset is composed of two columns, Date and Profit/Losses. 
#(Thankfully, your company has rather lax standards for accounting, so the records are simple.)

##TODO

#Import the pathlib and csv library

from pathlib import Path
import csv

# Set the file path
csvpath = Path('../PyBank/budget_data.csv')

# Initialize list of records
records = []

# Initialize line_num variable
line_num = 0

# Initialize variables
change_count = 0
total_change = 0
greatest_increase = 0
greatest_increase_month = None
previous_profit_losses = None
greatest_decrease = 0
greatest_decrease_month = None

# Open the csv file as an object
with open(csvpath, 'r') as csvfile:
    
    # Print the datatype of the file object
    print(f"Financial Analysis")
    print(f"----------------------------")
    
    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=',')
    
    # Read each row of data after the header
    header = next(csvreader)
    
    # The total number of months included in the dataset
    total_months = 0 
    net_total = 0
    for row in csvreader:
        months = row[0]
        profit_losses = int(row[1])
        total_months += 1
        net_total += profit_losses
        
    #The average of the changes in Profit/Losses over the entire period
        if previous_profit_losses is not None:
            change = profit_losses - previous_profit_losses
            total_change += change
            change_count += 1

    #The greatest increase in profits (date and amount) over the entire period
            if change > greatest_increase:
                greatest_increase = change
                greatest_increase_month = months
            
    #The greatest decrease in losses (date and amount) over the entire period
            if change < greatest_decrease:
                greatest_decrease = change
                greatest_decrease_month = months
        previous_profit_losses = profit_losses    

if change_count > 0:
    average_change = total_change / change_count
else:
    average_change = 0
average_change = round(average_change, 2)
        
print("Total Months:", total_months)       
print("Total: $", net_total)
print("Average Change: $%.2f" % average_change)
print("Greatest Increase in Profits:", greatest_increase_month, "($", greatest_increase, ")")
print("Greatest Decrease in Profits:", greatest_decrease_month, "($", greatest_decrease, ")")
       
# Set the path for the output.csv
output = Path("output.csv")

# Store the results in a list of lists
results = [
    ["Total Months", total_months],
    ["Total", net_total],
    ["Average Change", average_change],
    ["Greatest Increase in Profits", greatest_increase_month, greatest_increase],
    ["Greatest Decrease in Profits", greatest_decrease_month, greatest_decrease]
]

# Open the output path as a file and pass into the 'csv.writer()' function
# Set the delimiter/separater as a ','
with open(output, 'w') as csvfile:
    csvwriter = csv.writer(csvfile, delimiter=',')

    # Loop through the list of records and write every record to the
    # output csv file
    for row in results:
        csvwriter.writerow(row)
```
